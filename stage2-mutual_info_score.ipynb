# MUTUAL INFO SCORE 

```python
import pandas as pd
from sklearn import metrics

df=pd.read_excel('G:/redcarpet/datamodified.xls')
df1=pd.read_excel("G:/redcarpet/target.xlsx")
z=df.columns.values
#print(z[255])
t=[]
s=[]
for i in range(0,256):
    #print(i)
    if(i!=6 and i!=7 and i!=8):
        t.append(metrics.mutual_info_score(list(df[z[i]]),list(df1['target'])))
        s.append(z[i])

print(t)

x=pd.DataFrame()

x["non_funding variables"] = list(s)
x['mutual_info_score']=list(t)
x.to_csv("G:/redcarpet/mutual_info_score.csv")
print("done")
```

# sorting mutual_info_score

```python
df=pd.read_csv('G:/redcarpet/mutual_info_score.csv')

df1 = df.sort_values(by=["mutual_info_score"], ascending=False)

df1.to_csv("G:/redcarpet/mutual_info_score_sorted.csv")
print("done")
df1["mutual_info_score"]
```
